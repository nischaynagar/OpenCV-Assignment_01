# Importing an Image & Viewing it

```
from google.colab import drive
drive.mount('/content/drive')
```

```
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/drive/MyDrive/elonmusk.jpg")
cv2_imshow(image)
```

```
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2_imshow(rgb_image)
```

# Cropping

```
cropped = image[10:200, 100:2000]
cv2_imshow(cropped)
```

# Resizing

```
scale_percent = 10 # percent of original size
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
dim = (width, height)
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
cv2_imshow(resized)
#viewImage(resized, "After resizing with 20%")
```

# Rotating

```
(h, w, d) = image.shape
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 120, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
cv2_imshow(rotated)
```

# Grayscaling and Thresholding (Black & White effect)

```
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```
ret, threshold_image = cv2.threshold(gray_image, 150, 150, 0)
cv2_imshow(threshold_image)
```
